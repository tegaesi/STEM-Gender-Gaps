# Visualizing Gender Gap in STEM Fields, 1970-2012

### Introduction
The gender gap in STEM (which stands for science, technology, engineering, and mathematics) fields is reported on often in the news, but not everyone agrees that there is a gap. This guided project is an attempt to tell the gender gap story using line charts in matplotlib.

The U.S. Department of Education Statistics releases a dataset annually, containing the percentage of bachelor's degrees granted to women. This data set is broken up into 17 categories of degrees, with each column as a separate category, and covers the period of 1970-2012. Randal Olson, a data scientist at University of Pennsylvania, has cleaned the data set and made it available on his personal website.

# SECTION A: GENERAL EXPLORATION

### Importing Relevant Libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
```

### Loading the Dataset

```python
women_degrees = pd.read_csv("percent-bachelors-degrees-women-usa.csv")
women_degrees.head()
```

```python
women_degrees.info()
```

# SECTION B: VISUALIZATION

## Gender Gap in Biology

```python
fig = plt.figure()
plt.plot(women_degrees["Year"], women_degrees["Biology"])
plt.title("Percentage of Biology Degrees Awarded to Women ")
plt.show()
```

```python
fig = plt.figure()
plt.plot(women_degrees["Year"], women_degrees["Biology"], label= "Women")
plt.plot(women_degrees["Year"], 100 - women_degrees["Biology"], label= "Men")
plt.legend(loc="upper right")
plt.title("Percentage of Biology Degrees Awarded by Gender")
plt.show()
```

### Removing Tick Marks and Hiding Spines

To improve the data-ink ratio, let's make the following changes to the plot we created in the last step:

Remove all of the axis tick marks.
Hide the spines, which are the lines that connects the tick marks, on each axis.

```python
fig,ax = plt.subplots()

ax.plot(women_degrees["Year"], women_degrees["Biology"], label= "Women", c="Blue")
ax.plot(women_degrees["Year"], 100 - women_degrees["Biology"], label= "Men", c="Green")

ax.tick_params(bottom=False, top=False, left=False, right=False)
ax.spines["left"].set_visible(False)
ax.spines["bottom"].set_visible(False)
ax.spines["top"].set_visible(False)
ax.spines["right"].set_visible(False)

ax.legend(loc="upper right")
ax.set_title("Percentage of Biology Degrees Awarded by Gender")
plt.show()
```

## Comparing Gender Gaps Across Degree Categories

```python
major_cats = ["Biology", "Computer Science", "Engineering", "Math and Statistics"]
fig,ax = plt.subplots(figsize=(15,15))

for sp in range(0,4):
    ax = fig.add_subplot(2,2,sp+1)
    ax.plot(women_degrees["Year"], women_degrees[major_cats[sp]], label= "Women", c="Orange")
    ax.plot(women_degrees["Year"], 100 - women_degrees[major_cats[sp]], label= "Men", c="Blue")
    ax.set_xlim(1968,2011)
    ax.set_ylim(0,100)
   
    ax.tick_params(bottom=False, top=False, left=False, right=False)
    for key, spine in ax.spines.items():
        spine.set_visible(False)
    
    ax.set_title(major_cats[sp])

plt.legend(loc="upper right")
plt.show()
```

Few seconds reading the chart would reveal that Computer Science and Engineering have big gender gaps. While the gapa in Biology and Math and Statistics is quite small. In addition, the first two degree categories are dominated by men while the latter degree categories are much more balanced.

### Setting Line Colour Using RGB & Setting Line Width

```python
major_cats = ["Biology", "Computer Science", "Engineering", "Math and Statistics"]
fig = plt.figure(figsize=(15,15))

for sp in range(0,4):
    ax = fig.add_subplot(2,2,sp+1)
    cb_1 = (164/255,107/255,0/255)    #specifying a colour using RGB
    cb_2 = (14/255,255/255,128/255)
    
    ax.plot(women_degrees["Year"], women_degrees[major_cats[sp]], label= "Women", c=cb_1, linewidth=3)
    ax.plot(women_degrees["Year"], 100 - women_degrees[major_cats[sp]], label= "Men", c=cb_2, linewidth=3)
    ax.set_xlim(1968,2011)
    ax.set_ylim(0,100)
   
    ax.tick_params(bottom=False, top=False, left=False, right=False)
    for key, spine in ax.spines.items():
        spine.set_visible(False)
    
    ax.set_title(major_cats[sp])

plt.legend(loc="upper right")
plt.show()
```

### Improving Chart Layout

```python
stem_cats = ['Engineering', 'Computer Science', 'Psychology', 'Biology', 'Physical Sciences', 'Math and Statistics']

fig = plt.figure(figsize=(22, 3))

for sp in range(0,6):
    ax = fig.add_subplot(1,6,sp+1)
    ax.plot(women_degrees['Year'], women_degrees[stem_cats[sp]], c="blue", label='Women', linewidth=3)
    ax.plot(women_degrees['Year'], 100-women_degrees[stem_cats[sp]], c="orange", label='Men', linewidth=3)
    for key,spine in ax.spines.items():
        spine.set_visible(False)
    ax.set_xlim(1968, 2011)
    ax.set_ylim(0,100)
    ax.set_title(stem_cats[sp])
    ax.tick_params(bottom=False, top=False, left=False, right=False)

plt.legend(loc='upper right')
plt.show()
```

### Annotating: Replacing Legends with Line Charts Labelled Directly

```python
stem_cats = ['Engineering', 'Computer Science', 'Psychology', 'Biology', 'Physical Sciences', 'Math and Statistics']

fig = plt.figure(figsize=(22, 3))

for sp in range(0,6):
    ax = fig.add_subplot(1,6,sp+1)
    ax.plot(women_degrees['Year'], women_degrees[stem_cats[sp]], c="blue", linewidth=3)
    ax.plot(women_degrees['Year'], 100-women_degrees[stem_cats[sp]], c="orange", linewidth=3)
    for key,spine in ax.spines.items():
        spine.set_visible(False)
    ax.set_xlim(1968, 2011)
    ax.set_ylim(0,100)
    ax.set_title(stem_cats[sp])
    ax.tick_params(bottom=False, top=False, left=False, right=False)

    if sp==0:
        ax.text(2005, 87, "Men")
        ax.text(2002, 8, "Women")
    elif sp==5:
        ax.text(2005, 62, "Men")
        ax.text(2001, 35, "Women")

plt.show()
```

## Comparing Gender Gaps Across All Degree Categories

```python
fig = plt.figure(figsize=(16,20))   #creating the canvas on which all subplots are added; figsize=(width,hgt)

stem_cats = ['Psychology', 'Biology', 'Math and Statistics', 'Physical Sciences', 'Computer Science', 'Engineering']
for sp in range(0,18,3):
    cat_index = int(sp/3)
    ax = fig.add_subplot(6,3,sp+1)          #add subplots to the first column
    ax.plot(women_degrees["Year"], women_degrees[stem_cats[cat_index]], c="orange", linewidth=3)
    ax.plot(women_degrees["Year"], 100 - women_degrees[stem_cats[cat_index]], c="blue", linewidth=3)
    
    ax.tick_params(top=False, bottom=False, left=False, right=False, labelbottom=False) #remove all tick marks and the bottom tick labels
    ax.set_yticks([0,100])                  #label left tick with only 0 and 100
    for key,spine in ax.spines.items():     #loop through spines for each subplot and
        spine.set_visible(False)            #remove chart spines
        
    ax.axhline(50, c=(171/255, 171/255, 171/255), alpha=0.3)    #draw horizontal line at mark 50 with 0.3 transparency; alpha= 0 to 1 range
    ax.set_title(stem_cats[cat_index])      #title
    ax.set_xlim(1968,2011)                  #set x-limit
    ax.set_ylim(0,100)                      #set y-limit
    
    if cat_index==0:                        #annotate first subplot: ax.text(x-position, y-position, "text")
        ax.text(2005, 10, "Men")
        ax.text(2003, 85, "Women")
    elif cat_index==5:                      #annotate last subplot
        ax.text(2005, 87, "Men")
        ax.text(2003, 7, "Women")
        ax.tick_params(labelbottom=True)    #place bottom tick labels for the last graph
        

lib_arts_cats = ['Foreign Languages', 'English', 'Communications and Journalism', 'Art and Performance', 'Social Sciences and History']
for sp in range(1,16,3):
    cat_index = int((sp-1)/3)
    ax = fig.add_subplot(6,3,sp+1)          
    ax.plot(women_degrees["Year"], women_degrees[lib_arts_cats[cat_index]], c="orange", linewidth=3)
    ax.plot(women_degrees["Year"], 100 - women_degrees[lib_arts_cats[cat_index]], c="blue", linewidth=3)
    
    ax.tick_params(top=False, bottom=False, left=False, right=False, labelbottom=False)
    ax.set_yticks([0,100])
    for key,spine in ax.spines.items():
        spine.set_visible(False)
        
    ax.axhline(50, c=(171/255, 171/255, 171/255), alpha=0.3)
    ax.set_title(lib_arts_cats[cat_index])
    ax.set_xlim(1968,2011)
    ax.set_ylim(0,100)
    
    if cat_index==0:
        ax.text(2005, 18, "Men")
        ax.text(2003, 78, "Women")
    if cat_index==4:
        ax.tick_params(labelbottom=True)


other_cats = ['Health Professions', 'Public Administration', 'Education', 'Agriculture','Business', 'Architecture']
for sp in range(2,20,3):
    cat_index = int((sp-2)/3)    
    ax = fig.add_subplot(6,3,sp+1)          
    ax.plot(women_degrees["Year"], women_degrees[other_cats[cat_index]], c="orange", linewidth=3)
    ax.plot(women_degrees["Year"], 100 - women_degrees[other_cats[cat_index]], c="blue", linewidth=3)
    
    ax.tick_params(top=False, bottom=False, left=False, right=False, labelbottom=False)
    ax.set_yticks([0,100])
    for key,spine in ax.spines.items():
        spine.set_visible(False)
        
    ax.axhline(50, c=(171/255, 171/255, 171/255), alpha=0.3)
    ax.set_title(other_cats[cat_index])
    ax.set_xlim(1968,2011)
    ax.set_ylim(0,100)
    
    if cat_index==0:
        ax.text(2005, 5, "Men")
        ax.text(2003, 90, "Women")
    elif cat_index==5:
        ax.text(2005, 62, "Men")
        ax.text(2003, 30, "Women")
        ax.tick_params(labelbottom=True)

plt.savefig("gender degrees.png")           #saving as png file and must be called before plt.show(); alternatively use fig.savefig()
plt.show()
```

### Conclusion
The 17 charts above represent the 17 degrees under study. 

By this analysis, they can be grouped into three: fields where the gender gap is bridged (where the male to female ratio is 50:50), where the gap is small, and where the gap remains wide.

- Bridged Gap - Agriculture, Business, Social Sciences & History

- Small Gap - Architecture, Physical Sciences, Art & Performance, Biology, Maths & Statistics, Communications & Journalism, English

- Wide Gap - Psychology, Foreign Languages, Health Professions, Public Administration, Education -- far more women; Computer Science, Engineering -- far more men

Furthermore, the period surveyed showed five out of the eight STEM fields had more men graduating than women: Engineering, Architecture, Computer Science, Physical Sciences, Math & Statistics. Only in Biology, Health Professions and Agriculture were more women graduates found. 

Is it safe to conclude that women are more inclined towards life and nature than towards machines and man-made systems?
